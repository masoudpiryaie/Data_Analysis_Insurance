```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv('dataset.csv')
df.head()
```

## Data Preparation and Preprocessing
We inspect the dataset, check for missing values, and encode categorical variables.

**Note:** This step ensures data quality and prepares it for analysis.

```python
print(df.info())
print(df.describe())

# Encode categorical variables
df = pd.get_dummies(df, columns=['sex', 'region'], drop_first=True)
df['smoker'] = df['smoker'].map({'yes': 1, 'no': 0})
df.head()
```

## Exploratory Data Analysis (EDA)
**Note:** In this section, we analyze the data visually and statistically to answer the business questions.

### 1. Impact of Smoking on Insurance Premiums
**Note:** Smokers are expected to have significantly higher charges due to higher health risks.

```python
smoker_vs_non_smoker = df.groupby('smoker')['charges'].mean()
print(smoker_vs_non_smoker)
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Charges Distribution by Smoking Status')
plt.show()
```

### 2. Relationship between BMI and Premiums
**Note:** A higher BMI often correlates with higher charges, especially for smokers.

```python
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df, alpha=0.7)
plt.title('BMI vs Charges')
plt.show()
print("Correlation between BMI and Charges:", df['bmi'].corr(df['charges']))
```

### 3. Regional Differences in Premiums
**Note:** Regional averages help identify geographical patterns in costs.

```python
region_charges = df.groupby(['region_northwest','region_southeast','region_southwest']).mean()['charges']
print(region_charges)
sns.barplot(x='region_southeast', y='charges', data=df)
plt.title('Regional Impact on Charges')
plt.show()
```

### 4. Effect of Children on Premiums
**Note:** More dependents typically increase insurance costs.

```python
sns.barplot(x='children', y='charges', data=df, estimator=np.mean)
plt.title('Average Charges by Number of Children')
plt.show()
```

### 5. Characteristics of High-Risk Customers
**Note:** High-risk individuals usually combine smoking with high BMI and older age.

```python
top_10 = df.sort_values(by='charges', ascending=False).head(int(len(df)*0.1))
print(top_10.describe())
sns.pairplot(top_10[['age','bmi','charges','smoker']], hue='smoker')
plt.show()
```

### 6. Correlation Heatmap
**Note:** The heatmap shows how variables are related to each other numerically.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Conclusion & Recommendations  

The exploratory data analysis highlights several important insights into the factors driving medical insurance costs. The results clearly show that **smoking status, body mass index (BMI), and age** are the strongest predictors of high charges. Smokers, in particular, represent a disproportionately high financial risk, with their average charges being several times higher than non-smokers. Similarly, customers with higher BMI and older age groups exhibit a consistent increase in medical expenses, reflecting well-known health risk patterns.  

Beyond these dominant factors, **family size (number of children)** and **regional differences** also influence charges, though to a lesser degree. Families with more children tend to have slightly higher costs, which may be attributed to additional dependents covered under the policy. Regional variations, while not extreme, suggest the presence of socio-economic or healthcare infrastructure differences that could be leveraged for localized pricing strategies.  

### Recommendations  
1. **Refine Pricing Models:** Incorporate smoking, BMI, and age as core features in risk-based pricing models to ensure premiums reflect actual customer risk.  
2. **Targeted Health Programs:** Introduce wellness initiatives such as smoking cessation support and weight management programs. These can both reduce long-term claims and enhance customer loyalty.  
3. **Family-Oriented Policies:** Develop family-oriented insurance products with tailored pricing to balance affordability for larger households with risk exposure.  
4. **Regional Adjustments:** Consider slight premium adjustments by region to account for localized cost patterns and maintain competitiveness.  
5. **Enhanced Data Collection:** Expand data gathering efforts to include lifestyle and medical history variables (e.g., diet, exercise habits, chronic conditions) to improve predictive accuracy and refine future analyses.  

**Note:** By combining these insights with proactive health management initiatives, the company can not only mitigate financial risks but also strengthen its market position through fair, transparent, and data-driven pricing strategies.
